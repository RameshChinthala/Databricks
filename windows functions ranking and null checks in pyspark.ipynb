```python
#Types of Window Functions:
#1) Analytical Function: e.g., lead(), lag(), cume_dist()
#2) Ranking Function: e.g., row_number(), rank(), dense_rank(), percent_rank()
#3) Aggregate Function: e.g., avg(), sum(), min(), max()

from pyspark.sql.window import Window
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("pysaprk_window").getOrCreate()
sampleData = (("Olivia", 28, "Sales", 3000,'12-feb-2022'),
              ("Harry", 33, "Sales", 4600,'13-feb-2022'),
              ("Smith", 40, "Sales", 4100,'14-feb-2022'),
              ("Marry", 25, "Finance", 3000,'12-feb-2022'),
              ("Henry", 28, "Sales", 3000,'15-feb-2022'),
              ("Lars", 46, "Management", 3300,'17-feb-2022'),
              ("Jeny", 26, "Finance", 3900,'18-feb-2022'),
              ("Aya", 30, "Marketing", 3000,'19-feb-2022'),
              ("Omar", 29, "Marketing", 2000,'20-feb-2022'),
              ("Johnny", 39, "Sales", 4100,'22-feb-2022')
              )

columns = ["Employee_Name", "Age", "Department", "Salary", "Join_Date"]

df = spark.createDataFrame(data= sampleData, schema=columns)

windowPartition = Window.partitionBy("Department").orderBy("Age")
df.printSchema()

df.show()
```

```python
df.show()

from pyspark.sql.functions import col, dense_rank

# Find the second highest distinct salary
df_with_rank = df.withColumn(
    "rank",
    dense_rank().over(Window.orderBy(col("salary").desc()))
)
second_highest_salary = df_with_rank.filter(col("rank") == 2).select("salary").distinct()
display(second_highest_salary)


# Count nulls for all columns, apply isnan only to numeric columns
numeric_types = (DoubleType, FloatType, IntegerType, LongType)
df.select([
    count(when(
        (col(c).isNull()) | (isnan(c) if isinstance(df.schema[c].dataType, numeric_types) else False),
        c
    )).alias(c) for c in df.columns
]).show()




```
