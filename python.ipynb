## **Slow Query** 

```python
#Step 1 Avoid SELECT * 
# Step-2 Partition the Table
Partition by sale_date because it is used in the filter.
CREATE TABLE large_table
(
sale_id INT,
customer_id INT,
product_id INT,
sale_amount DECIMAL(10,2),
sale_date DATE
)
USING DELTA
PARTITIONED BY (sale_date)

# stpe 3
Use ZORDER

Within a partition, organize data by customer_id.
OPTIMIZE large_table
ZORDER BY (customer_id);

#step 4

✔ Partition → sale_date
✔ ZORDER → customer_id
✔ OPTIMIZE → File compaction
✔ Cache if repeated queries
```
