The Show() function in pyspark

|Scenario | Code | Output |
|---------|------|--------|
|Show all Data(default) | df.show() | shows 20 rows , trunctes long string |
| show limited rows | df.show(5) | disply only the first 5 rows |
| Truncate long text(default 20 chars) | df.show(truncates = true) | cuts of long text |
|show full text | df.show(truncate = Falese) |  display full text|
| show vertical format | df.show(vertical = true) | display data one at a time |
| set customer truncate length| df.show(truncate=10) | cuts off text 10 char |
