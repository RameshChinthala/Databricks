## How to Use withColumn() Function in PySpark | Add & Update & distinct Columns

```python
from pyspark.sql.functions import col, lit, expr

 #Syntax
#DataFrame.withColumn(colName, col)
# Data Frame is immutable , meaning we can't modifiy an exsiting data frame in place using withcolumn using this we can't modify the orginal dataframe insted of we can create new dataframs with adding or modifiying
data = [
      (1, "Alice", 5000, "it"),
    (2, "Bob", 6000, "Finace"),
    (3, "Charlie", 7000, "HR"),
    (4, "David", 8000, "casa"),
    (5, "Eve", 9000, "it"),
    (6, "Frank", 10000, "casa"),
    (7, "Grace", 11000, "HR"),
    (8, "Hannah", 12000, "Finace"),
    (9, "Ian", 13000, "casa"),
    (10, "Jack", 14000, "HR")
]

df = spark.createDataFrame(data, ["id", "name", "salary", "department"])
df.show()


#Add New Column
df_new = df.withColumn("bonus", lit(1000))
df_new.show()

#Update Existing Column
df_updated = df_new.withColumn("salary", col("salary") * 1.10)
df_updated.show()

#Use Expressions with expr()
df_expr = df_updated.withColumn("salary_with_bonus", expr("salary + bonus"))
df_expr.show()

# Change Column Data Type
df_type_changed = df_expr.withColumn("salary", col("salary").cast("Integer"))
df_type_changed.show()

#Rename a Column
df_renamed = df_type_changed.withColumn("emp_name", col("name")).drop("name")
df_renamed.show()


#Limit a Column
df = df.limit(5).show()



```
