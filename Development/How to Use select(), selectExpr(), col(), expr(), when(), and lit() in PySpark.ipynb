
## How to Use select(), selectExpr(), col(), expr(), when(), and lit() in PySpark

select() - Retrieve specific columns.
selectExpr() - Use SQL expressions.
col() - Reference columns.
expr() - Perform expressions.
when() - Conditional logic.
lit() - Add constant columns.
orderBy() - sort -True for ascending (default), False for descending


```python
from pyspark.sql.functions import col, expr, when, lit

data = [
  (1, "Aamir Shahzad", "IT", 6000, 22),
    (2, "Ali", "HR", 7000, 26),
    (3, "Raza", "Finance", 8000, 29),
    (4, "Aamir Shahzad", "IT", 6000, 33),  # Duplicate Entry
    (5, "Alice", "IT", 5000, 21),
    (6, "Bob", "HR", 7000, 32),
    (7, "Ali", "HR", 7000, 22),            # Duplicate Entry
    (8, "Charlie", "Finance", 9000, 72),
    (9, "David", "IT", 6000, 27),
    (10, "Eve", "HR", 7500, 26)
]


df = spark.createDataFrame(data, ["id", "Name", "dept", "salary",  "age"]                           )
df.show()                           

# select specific columns
df.select("id","Name").show()

#selecting column using col()
df.select(col("id"),col("salary")).show()

#renaming columns using alias()
df.select(col("id").alias("Employee ID") , col("Name").alias("Employee Name")).show()

#selecting columns using expr()
df.select("name", "salary", expr("salary * 1.10 AS increased_salary")).show()

#Using Conditional Expressions with when()
df.select(
    "name",
    "salary",
    when(col("salary") > 7000, "High").otherwise("Low").alias("Salary_Category")
).show()

# Using selectExpr() for SQL-like Expressions
df.selectExpr("name", "salary * 2 as double_salary").show()

# Adding Constant Columns Using lit()
df.select("name", "dept", lit("Active").alias("status")).show()

# Selecting Columns Dynamically
columns_to_select = ["name", "salary", "dept"]
df.select(*columns_to_select).show()

#) select all columns expect one
df.select([col for col in df.columns if col != "salary"]).show()



# Syntax
# DataFrame.orderBy(*cols, ascending=True)

#The orderBy() function in PySpark is a powerful tool to sort data efficiently. It can handle single or multiple columns and allows flexibility with ascending or descending order settings.

# Order by Single Column (Ascending - Default)

df.orderBy(col("salary").desc()).show()

df.orderBy("dept", "salary").show()

# Order by Multiple Columns with Mixed Order
df.orderBy(col("dept").asc(), col("salary").desc()).show()

# Using orderBy() with Explicit Boolean for Ascending/Descending
df.orderBy(col("salary"), ascending=True).show()
df.orderBy(col("salary"), ascending=False).show()


#Removing Duplicate Rows using distinct()
df_distinct = df.distinct()
df_distinct.show()

#Getting Unique Values from a Single Column
df.select("dept").distinct().show()


```

## The dropDuplicates() 
function in PySpark is used to remove duplicate rows from a DataFrame. It returns a new DataFrame with only unique rows, keeping the first occurrence of each duplicate


```python
#drop the columns 
#df.drop("dept","age").show()

df.show()

# Removing duplicate rows (entire row must match)
df_no_duplicates = df.dropDuplicates()
df_no_duplicates.show()

# Removing duplicates based on 'name' and 'department'
df_no_duplicates_specific = df.dropDuplicates(["name", "dept"])
df_no_duplicates_specific.show()
```
