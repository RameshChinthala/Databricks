```python

from pyspark.sql import SparkSession
from pyspark.sql import Row

# Initialize SparkSession
spark = SparkSession.builder.appName("DropNAExample").getOrCreate()

# Sample data
data = [
  (1, "Aamir Shahzad", "IT", None, 22),
    (2, "Ali", "HR", 7000, 26),
    (3, "Raza", "None", 8000, 29),
    (4, "Aamir Shahzad", "IT", 6000, 33),  # Duplicate Entry
    (5, "Alice", "IT", 5000, 21),
    (None, "Bob", "HR", 7000, 32),
    (7, "Ali", "HR", 7000, 22),            # Duplicate Entry
    (8, "Charlie", "Finance", 9000, 72),
    (9, "David", "IT", 6000, 27),
    (10, "Eve", "HR", 7500, 26),
    (None,None,None,None,None)
]


df = spark.createDataFrame(data, ["id", "Name", "dept", "salary",  "age"]                           )
df.show() 

# Removing duplicate rows (entire row must match)
df_no_duplicates = df.dropDuplicates()
df_no_duplicates.show()

# Removing duplicates based on 'name' and 'department'
df_no_duplicates_specific = df.dropDuplicates(["name", "dept"])
df_no_duplicates_specific.show()

# Remove rows with any NULL values
df_clean_any = df.dropna()
df_clean_any.show()

#fillna - Replace NULL values with specified values
df_fillna = df.fillna({"id": 0, "Name": "Unknown", "dept": "Unknown", "salary": 0})
df.show()

#filter - Filter rows based on a condition
df_filtered = df.filter(df.age > 25)
df_filtered.show()

df_filtered = df.filter((df.salary == 7000) & (df.age < 30)) # for or use |
df_filtered.show()

#group by
df_groupby = df.groupBy("dept")\
               .agg({"salary": "avg"})
df_groupby.show()

#orderBy
df_orderby = df.orderBy("salary")
df_orderby.show()

#sort
df_sort = df.sort("salary")
df_sort.show()

#sortWithinPartitions
df
```

**Basic Syntax**
DataFrame.dropna(how='any', thresh=None, subset=None)

**how:** 'any' (default) drops a row if any nulls are present. 'all' drops a row only if all columns are null.

**thresh:** Minimum number of non-null values required to retain the row.

**subset:** List of column names to consider for null checking.

```python
df.show()
# Remove rows with any NULL values
df_clean_any = df.dropna()
df_clean_any.show()


# Remove rows where all columns are NULL
df_clean_all = df.dropna(how='all')
df_clean_all.show()

# Remove rows with NULLs in 'name' or 'salary'
df_clean_subset = df.dropna(subset=['id','name', 'salary'])
df_clean_subset.show()
```

```python
df.show()

#count() - Total Number of Rows (Including Duplicates)
total_rows = df.count()
print("Total number of rows:", total_rows)

#distinct().count() - Counts Unique Rows
distinct_rows = df.distinct().count()
print("Number of distinct rows:", distinct_rows)

#groupBy() + count() - Count Occurrences of Each Name
df.groupBy("Name").count().show()
      
```
