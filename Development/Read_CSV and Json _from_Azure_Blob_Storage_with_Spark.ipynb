## How to Read CSV files from Azure Could 

```python
storage_account_name = "enter Name"
contaner_name = "enter container name"
storage_account_access_key = "enter key" # a shared access signature copy from the storage account o file right side click then generate sas token

# wasbs is the portocal enbales secure to azure blob storage in databricks or spark
filepath = "wasbs://"+contaner_name+"@"+storage_account_name+".blob.core.windows.net/sample_data/sample_data.csv"
# configure spark to use sas token authontication 
spark.conf.set(
  "fs.azure.account.key."+storage_account_name+".blob.core.windows.net",
  storage_account_access_key)

df = spark.read.format("csv")\
               .option("header"="True")\
               .option("interSchema" = "True")\
               .load(filepath)
df.show()

```

## How to read Json files from azure blob

```python
storage_account_name = "enter Name"
contaner_name = "enter container name"
storage_account_access_key = "enter key" # a shared access signature copy from the storage account o file right side click then generate sas token

# wasbs is the portocal enbales secure to azure blob storage in databricks or spark
filepath = "wasbs://"+contaner_name+"@"+storage_account_name+".blob.core.windows.net/sample_data/sample_data.json"
# configure spark to use sas token authontication 
spark.conf.set(
  "fs.azure.account.key."+storage_account_name+".blob.core.windows.net",
  storage_account_access_key)

#multiline is used to read json file with multiple lines other wise no need to meantion
df = spark.read.option("multiline","true")\
               .json(filepath)
df.display()

df.printschema()

df.select("id",
            "first_name").show(truncate=False)

```
