## What is Data Frame in Pyspark?

A dataframe in pysaprk is table like structure that organizes data into rows and columns, just like sql, excel , table or pandas data frame. 
it help to prrocess big data efficiently by using mutiple computters at the same time.

- Looks like table : Has columns with names and specifiy data type.
- fats and scalble : works well with large datasets.
- works with diff sources : can read csv, json , parquet ,databases, etc
- allows sql quaries : you can use sql like commands to filter and group data.


## Create the Dataframe from a list of tuple(static)[](url)

```python
from pyspark.sql import Row

data = [
  Row(id=1 , Name = "Ramesh" , age=33), #Tuple 1
  Row(id=2, Name = "rakesh" , age=34), #Tuple 2
  Row(id=3 , Name = "bhaga" ,  age=35) #Tuple 3
]

#create a dataFram from the list

df = spark.createDataFrame(data)
df.show()

'''A Tuble in python is ordered, immutable sequence of element. "Immutable" means that once a tuple is created its element cannot changed.
Tuble are defined by enclosing element within paranthese(), separated by commas.'''
```

## Create DataFrame using with list of Dictionaries

```python
#A Dictionary list is a list of dictionaries , where each dictionary represents a recods with key -value pairs. this structure is useful for storing structure data. simlier like table
data = [
  {"id" : 1 , "Name": "Ramesh" , "age":33}, #Dictionary 1
  {"id" : 2 , "Name": "rk" , "age":33}, #Dictionary 2
  {"id" : 3 , "Name": "sk" , "age":45} #Dictionary 3
]
df1 = spark.createDataFrame(data)
df1.show()
```

```python
from pyspark.sql.functions import explode

data = [{
    "id": "123",
    "name": "test_name",
    "dept": "D&A",
    "address": {"city": "pune", "pin": "411013"}, #is treated like a key-value pair, which Spark does NOT understand Use dot notation for struct

    "salary": [
        {"salamount": "2000", "startdate": 20210101},
        {"salamount": "2500", "startdate": 20220101} #Use explode() for arrays
    ]
}]

df3 = spark.createDataFrame(data)
df_flat = df3.select(
    "id",
    "name",
    "dept",
    "address.city",  
    "address.pin",
    explode("salary").alias("salary")
).select(
    "id",
    "name",
    "dept",
    "city",
    "pin",
    "salary.salamount",
    "salary.startdate"
)

display(df_flat)
```

## How to add column name to DataFrame and also check schema of dataFrame

toDF() - to rename the coolumn

```python
withoutheader_data = [
  (1, "Ramesh", 33),
  (2, "rakesh", 34),
  (3 ,"dell", 35)
]

df = spark.createDataFrame(withoutheader_data)
df.show()
#df = df.toDF("id", "Name", "age") #without header we added header  using toDF()
#df.show()

#comibing create data frame () and todf() in onestep

df = spark.createDataFrame(withoutheader_data).toDF("id", "Name", "age")
df.show()

#check with schema for  dataFrame
df.printSchema()
```

## below is another way to write the without header

```python
withoutheader_data = [
  (1, "Ramesh", 33),
  (2, "rakesh", 34),
  (3 ,"dell", 35)
]

df_without_header = spark.createDataFrame(withoutheader_data)
df_without_header.show()
df_without_header = spark.createDataFrame(withoutheader_data, schema = ["id", "Name", "age"])
df_without_header.show()
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField("id", StringType(), True),
    StructField("Name", StringType(), True),
    StructField("age", IntegerType(), True)
])

df_without_header = spark.createDataFrame(withoutheader_data, schema = schema)
df_without_header.show()
```
