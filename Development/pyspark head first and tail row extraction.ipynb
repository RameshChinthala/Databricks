**head()** : return the first n rows 

**first()** : return the very first rows

**tail()**: : return the last n rows

```python
from pyspark.sql import SparkSession

# Initialize SparkSession
spark = SparkSession.builder.appName("pysparkheadfirst").getOrCreate()

# Sample data
data = [
  (1, "Aamir Shahzad", "IT", None, 22),
    (2, "Ali", "HR", 7000, 26),
    (3, "Raza", "None", 8000, 29),
    (4, "Aamir Shahzad", "IT", 6000, 33),  # Duplicate Entry
    (5, "Alice", "IT", 5000, 21),
    (None, "Bob", "HR", 7000, 32),
    (7, "Ali", "HR", 7000, 22),            # Duplicate Entry
    (8, "Charlie", "Finance", 9000, 72),
    (9, "David", "IT", 6000, 27),
    (10, "Eve", "HR", 7500, 26),
    (None,None,None,None,None)
]
columns = ["id", "Name", "dept", "salary",  "age"]

df = spark.createDataFrame(data, columns   )
df.show() 

print("using head")

df_headrows = df.head(3)
for i in df_headrows:
   print(i)

print("using head")

df_first = df.first(3)


print("using tail")

dftail = df.tail(3)
for i in dftail:
   print(i)
```
